#### Importando Bibliotecas

```python
#Manipulação de Dados
import pandas as pd

#Requisição da API
import requests
from tmdbv3api import Movie, TMDb, Discover

#Tratamento do JSON
import json
```

```python
# Na variável "tmdb.api_key" digite sua key de api do TMDB. Para gera uma entre no site e crie uma conta.
tmdb_movie = Movie()
tmdb = TMDb()
tmdb.api_key = '5b23ce0eeb95b91e1fae905d65606841'
discover = Discover()
```

#### Funções aplicadas

```python
# Esta é uma função que utiliza da biblioteca que foi baseada na API do TMDB.
# Informe o nome de um filme que ela lhe trará o ID. 
def get_id_movie(title_movie):
  result = tmdb_movie.search(title_movie)
  movie_id = result[0].id
  return movie_id
```
